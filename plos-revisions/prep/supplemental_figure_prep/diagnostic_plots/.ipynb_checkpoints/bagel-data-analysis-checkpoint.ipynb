# Analysis of kinetic data for "over 100" mutants of ß-glucosidase B

## Size and shape of the set

The "Bagel" data set, as it has become known, comprises 2970 individual observations of ß-glucosidase B from P. polymxa acting on 4-nitrophenyl-ß-D-glucoside

```python
import pandas as pd
plates = pd.read_csv('plates.csv')
plates.head()
```

## Processing the raw data 

To begin, I wrote a script, `process.py`, to process the raw data. The script splits up the data by sample and attempts three fits on each sample: $$y=mx+b, \\ kobs=kcat*S/(km+S), \\ kobs=kcat*S/(km+S+ki)$$. 

`process.py` constructs a new data table, stored in `raw-out.csv`, keyed by sample name, and diagnostic plots for each sample.

```python
ro = pd.read_csv('out/raw-out.csv')
```

Let's add some useful metadata

```python
# metadata 
ro['seq'] = ro['sample'].str.replace(r'[a-z]', '')
ro['newaa'] = ro['sample'].str[-1]
ro['oldaa'] = ro['sample'].str[0]
```

```python
ro.describe()
```

```python
ro.head()
```

Processing with `process.py` reveals that we have 119 unique sample names in our data set, with an average yield of 0.7 ± 0.5 mg/mL and a maximum recorded yield of 2 mg/mL. 

## Lots and lots of errors

As you can see from the summary table, the average error for kcat is actually larger than the magnitude of the value. Clearly, we have come cleaning up to do. 

## Cleanup 

First things first, let's filter out all the wild types and average them.

```python
import re
wts = ro[(ro['sample'].str.contains('wt'))] 
wts
```

```python
mean_wts = wts.mean()
mean_wts.set_value('sample', 'wt')
mean_wts
```

So now that we have the mean values for the WT, let's remove all the individual WTs from the data set and add back in our average value.

```python
ro = ro[( ro['sample'].str.match('wt') == False )]
ro.loc[107] = mean_wts
ro.tail()
len(ro)
```

This leaves us with 106 mutants tested. We're already getting close to "over 100". 

## Low-yeilding mutants 

We'll also want to get rid of mutants that fail our yield cutoff, 0.165 mg/mL.

```python
noy = ro[(ro['yield'] < 0.165 )]
ro = ro[(ro['yield'] > 0.165 )]
len(noy)
```

Which gits rid of 14 mutants:

```python
noy
```

```python
noy.describe()
```

```python
len(ro)
```

## Linear fits

Probably a good place to start cleaning up is to determine what data fit best to straight lines. For this, let's say that anything with a linear r-square of greater than 0.95 is linear.

```python
linear = ro[(ro['R'] > 0.95)]
ro = ro[(ro['R'] < 0.95)]
len(linear)
```

That's another 12 mutants right there, let's take a look.

```python
linear
```

```python
linear.describe()
```

```python
linear['eff'] = linear['slope']
linear['err3'] = linear['std_err']
linear['kcat'] = linear['km'] = linear['err1'] = linear['err2'] = 'NaN'
```

```python
len(ro)
```

Right now, we're left with 80 mutants that will provide kcats and KMs. The ones that fail our error cutoff are

```python
high_err = ro[(ro['err1']/ro['kcat']>0.25)|(ro['err2']/ro['km']>0.25)]
ro = ro[(ro['err1']/ro['kcat']<0.25)|(ro['err2']/ro['km']<0.25)]
```

```python
high_err
```

```python
len(ro)
```

And we're left with 77 mutants that will provide kcats and KMs, and 12 that will provide a kcat/km value, for a total of 89. Let's calculate the kcat/KM for these too

```python
ro['eff'] = ro['kcat']/ro['km']
ro['err3'] = ro['kcat']/ro['km'] * ( (ro['err1']/ro['kcat'])**2 + (ro['err2']/ro['km'])**2 )**0.5

# logs and normalized logs
ro['log kcat'] = log(ro['kcat']+1)
ro['log km'] = log(ro['km']+1)
ro['normalized kcat'] = ro['log kcat'] / max(ro['log kcat'])
ro['normalized km'] = ro['log km'] / max(ro['log km'])
```

```python
final_set = ro.append(linear)
len(final_set)
```

```python
final_set.head()
```

Let's now break this up into the designed mutations and the alanine scan (using a crude metric)

```python
final_set['log eff'] = log(final_set['eff']+1)

final_set['normalized eff'] = final_set['log eff'] / max(final_set['log eff'])
ascan = final_set[ (final_set['sample'].str[-1] == 'a') ]
designs = final_set[ (final_set['sample'].str[-1] != 'a') ]
```
